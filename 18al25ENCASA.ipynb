##Ejercicio 18

FOR

```python
numero = int(input("Ingrese un número: "))

if numero < 0:
    print("No existen factoriales negativos")
else:
    factorial = 1
    for i in range(1, numero + 1):
        factorial = factorial * i

    print("El factorial de", numero, "es:", factorial)
```

WHILE

```python
numero = int(input("Ingrese un número: "))

if numero < 0:
    print("No existen factoriales negativos")
else:
    factorial = 1
    i = 1

    while i <= numero:
        factorial = factorial * i
        i += 1

    print("El factorial de", numero, "es:", factorial)
```

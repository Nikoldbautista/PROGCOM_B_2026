For


```python
numero = int(input("¿De qué número quieres la tabla?: "))
if numero <= 10:
 print(f"Tabla del {numero}:")
 for i in range(1, 10):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
else:
    print("Error: El número digitado es mayor a 10. No se puede mostrar la tabla.")
```

while

```python
numero = int(input("¿De qué número quieres la tabla?: "))
contador = 1
if numero <= 10:
  print(f"Tabla del {numero}")
  while contador <= 10:
    resultado = numero * contador
    print(f"{numero} x {contador} = {resultado}")
    contador += 1
else:
    print("Número no válido. Intenta de nuevo.")
```

```python
n = int(input("Introduce hasta qué número quieres sumar: "))
suma = 0

for i in range(1, n + 1):
    suma += i

print(f"La suma total es: {suma}")
```

While

```python
n = int(input("Introduce hasta qué número quieres sumar: "))
suma = 0
contador = 1

while contador <= n:
    suma += contador
    contador += 1

print(f"La suma total es: {suma}")
```

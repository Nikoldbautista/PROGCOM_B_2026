````python
figura=["angelitos","estrella fugaz","dinosaurio","diablitos"]#Creamos una lista por extencion
cantidad=[4,2,5,3]
````

#Operaciones con listas

4. Agregar elementos a la lista

###1. Recorrer listas

````python
len(figura) #len: Calcula el tamaño de una dato
````

````python
for elemento in range(len(figura)):
  print(figura[elemento])
````

````python
for elemento in range(len(figura)):
  print(figura[elemento], ":", cantidad[elemento])
````

````python
for elemento in figura:
  print(elemento)
````

##2. Verificar elemtentos dentro de la lista


````python
user=input("Que figura de galleta quiere comprar?")
for elemento in range(len(figura)):
  if user==figura[elemento]: print("Si hay galletas")
  else: print("no hay galleta")
````

````python
if user in figura: print("Si hay")
else: print("No hay")
````

````python
print("Si hay" if user in figura else "No hay")
````

```python
import math

a = float(input("Ingrese a: "))
b = float(input("Ingrese b: "))
c = float(input("Ingrese c: "))

d = b**2 - 4*a*c

if d > 0:
    x1 = (-b + math.sqrt(d)) / (2*a)
    x2 = (-b - math.sqrt(d)) / (2*a)
    print("Dos soluciones:", x1, "y", x2)

if d == 0:
    x = -b / (2*a)
    print("Una solución:", x)

if d < 0:
    print("No tiene soluciones reales")
```

```python
a = float(input("Lado 1: "))
b = float(input("Lado 2: "))
c = float(input("Lado 3: "))

es_triangulo = a + b > c and a + c > b and b + c > a
resultado = "No es un triángulo"

if es_triangulo and a == b and b == c: resultado = "Equilátero"
if es_triangulo and ((a == b and a != c) or (a == c and a != b) or (b == c and b != a)): resultado = "Isósceles"
if es_triangulo and a != b and a != c and b != c: resultado = "Escaleno"

print(f"Resultado: {resultado}")
```

IF ELIF

```python
a = float(input("Lado 1: "))
b = float(input("Lado 2: "))
c = float(input("Lado 3: "))

if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        tipo = "Equilátero"
    elif a == b or a == c or b == c:
        tipo = "Isósceles"
    else:
        tipo = "Escaleno"
    print(f"Es un triángulo {tipo}")
else:
    print("No es un triángulo")
```
